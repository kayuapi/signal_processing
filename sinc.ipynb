```python
# Copyright Tay Zhi Cong, year 2025

import numpy as np 
import matplotlib.pyplot as plt 
from scipy import signal 
import math
from ipywidgets import interact, FloatSlider
```

```python
dat = np.sinc(np.arange(-4, 4+0.1, 0.1))

plt.figure()
ax = plt.subplot(2,1,1)
ax.stem(np.arange(-4, 4+0.1, 0.1), dat)
ax.grid(which="major")
ax.grid(which="minor")
ax.minorticks_on()

ff = np.arange(-0.5, 0.5, 1/1024) * 10
f_dat = np.fft.fft(dat/sum(dat), 1024)
ax = plt.subplot(2,1,2)
ax.plot(ff, np.fft.fftshift(20*np.log10(abs(f_dat))))
ax.grid(which="major")
ax.grid(which="minor")
ax.minorticks_on()
ax.set_xlim(-5, 5)
ax.set_ylim(-60, 10)
```
