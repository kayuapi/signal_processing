```python
# Copyright Tay Zhi Cong, year 2025

import random
import numpy as np 
import matplotlib.pyplot as plt 
from scipy import signal 
import math
import sys
np.set_printoptions(threshold=sys.maxsize)
```

#### Design a low-pass half-band filter with the following spec:
transition bandwidth: 0.1 \
stop-band attenuation: 60dB

```python
stop_band_attenuation = 60
f_sampling = 1
f_transition = 0.1
f_minus_6db = 0.25

beta = signal.kaiser_beta(stop_band_attenuation) # 1. calculate the kaiser beta
filter_length = math.ceil((f_sampling/f_transition) * (stop_band_attenuation-8)/14) # 2. estimate the filter length : N ~= f_sampling/ f_transition * (Attenuation (in dB) - 8) / 14
print("beta is", beta)
print("filter_length is", filter_length)

filter_length = 41

filter_tap = signal.windows.kaiser(filter_length, beta=beta)
#print(filter_length)
h = np.sinc(2*f_minus_6db/f_sampling*np.linspace(-(filter_length - 1)/2,(filter_length - 1)/2,filter_length))*filter_tap
print("sum of h is", sum(h))
h = h/sum(h)
```

#### stem the impulse response, plot the magnitude spectrum and then the log magnitude spectrum

```python
ax1 = plt.subplot(3,1,1)
ax1.stem(h)
ff = np.arange(0, 1, 1/1024) * 1
ax2 = plt.subplot(3,1,2)
ax2.plot(ff, abs(np.fft.fft(h, 1024)))
ax2.grid(True)
ax2.minorticks_on()
ax2.grid(which="minor")
ax2 = plt.subplot(3,1,3)
ax2.plot(ff, 20*np.log10(abs(np.fft.fft(h, 1024))))
ax2.grid(True)
ax2.minorticks_on()
ax2.grid(which="minor")

plt.figure()
ax3 = plt.subplot(2,1,1)
ax3.plot(ff, 20*np.log10(abs(np.fft.fft(h, 1024))))
ax3.grid(True)
ax3.minorticks_on()
ax3.grid(which="minor")
ax3.set_xlim(0, 0.25)
ax3.set_ylim(-0.01, 0.01)

ax4 = plt.subplot(2,1,2)
ax4.plot(ff, 20*np.log10(abs(np.fft.fft(h, 1024))))
ax4.grid(True)
ax4.minorticks_on()
ax4.grid(which="minor")
ax4.set_xlim(0.28, 0.33)
ax4.set_ylim(-70, -50)
```

#### the locations of the impulse response zeros do not exist. They sit betwen impulse response sammple 1 and sample 2, sample 3 and smample 4, sample 5 and sample 6, sample 7 and sample 8...

#### Scale the impulse response and replot it

```python
h = h/sum(h)
ax1 = plt.subplot(3,1,1)
ax1.stem(h)
ff = np.arange(0, 1, 1/1024) * 1
ax2 = plt.subplot(3,1,2)
ax2.plot(ff, abs(np.fft.fft(h, 1024)))
ax2.grid(True)
ax2.minorticks_on()
ax2.grid(which="minor")
ax2 = plt.subplot(3,1,3)
ax2.plot(ff, 10*np.log10(abs(np.fft.fft(h, 1024))))
ax2.grid(True)
ax2.minorticks_on()
ax2.grid(which="minor")
```

####  the length of the filter to have alternate values of the impulse response be zero

To have alternate values of the impulse response be zero, the length of the filter must be odd.

In solution to 8_1, N is 21 and (N-1)/2 is even, therefore its first sample is zero.

In solution to 8_2, N is 23 and (N-1)/2 is odd, therefore its first sample is not zero.

Note that the first sample being zero is a desired property for a polyphase partition.
This is a desired property for a half-band filter and
for a polyphase partition
