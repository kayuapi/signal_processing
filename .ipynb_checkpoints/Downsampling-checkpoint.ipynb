```python
# Copyright Tay Zhi Cong, year 2025

import numpy as np 
import matplotlib.pyplot as plt 
from scipy import signal 
import math
from ipywidgets import interact, FloatSlider
```

```python
h1 = np.sinc(np.arange(-8, 8+0.1, 0.1)) * signal.windows.kaiser(161, beta=8)
h1 = h1/sum(h1)

freq_response = np.fft.fft(h1, 1024)
print(len(freq_response))
freq = np.arange(-0.5, 0.5, 1/len(freq_response)) * 2000

plt.figure()
ax = plt.subplot(2,1,1)
ax.plot(np.arange(0, 16+0.1, 0.1), h1)
ax.grid(which="major")
ax.grid(which="minor")
ax.minorticks_on()

ax = plt.subplot(2,1,2)
ax.plot(freq, np.fft.fftshift(20*np.log10(abs(freq_response))))
ax.grid(which="major")
ax.grid(which="minor")
ax.minorticks_on()
ax.set_xlim(-620, 620)
ax.set_ylim(-100, 1)

plt.figure()
ax = plt.subplot(2,1,2)
ax.plot(freq, np.fft.fftshift(20*np.log10(abs(freq_response))))
ax.grid(which="major")
ax.grid(which="minor")
ax.minorticks_on()
ax.set_xlim(80, 110)
ax.set_ylim(-10, 0)


plt.figure()
ax = plt.subplot(2,1,1)
ax.plot(freq, (np.unwrap(np.angle(freq_response))))
ax.grid(which="major")
ax.grid(which="minor")
ax.minorticks_on()
# ax.set_xlim(-0, 30)
# ax.set_ylim(-4, 4)
```
