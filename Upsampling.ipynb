```python
# Copyright Tay Zhi Cong, year 2025

import numpy as np 
import matplotlib.pyplot as plt 
from scipy import signal 
import math
```

```python
h1 = np.sinc(np.arange(-10, 10+0.25, 0.25)) * signal.windows.kaiser(81, beta=6)
h1 = h1/sum(h1)

freq_response = np.fft.fft(h1, 1024)
print(len(freq_response))
freq = np.arange(-0.5, 0.5, 1/len(freq_response)) * 400

plt.figure()
ax = plt.subplot(2,1,1)
ax.plot(np.arange(0, 20+0.25, 0.25), h1)
ax.grid(which="major")
ax.grid(which="minor")
ax.minorticks_on()

ax = plt.subplot(2,1,2)
ax.plot(freq, np.fft.fftshift(20*np.log10(abs(freq_response))))
ax.grid(which="major")
ax.grid(which="minor")
ax.minorticks_on()
ax.set_xlim(-620, 620)
ax.set_ylim(-100, 1)
```

```python
h2 = np.stack([h1, np.zeros(81)], axis=-1).reshape(162,)

freq_response = np.fft.fft(h2, 1024)
print(len(freq_response))
freq = np.arange(-0.5, 0.5, 1/len(freq_response)) * 800

plt.figure()
ax = plt.subplot(2,1,1)
ax.plot(np.arange(0, 40.25 + 0.25, 0.25), h2)
ax.grid(which="major")
ax.grid(which="minor")
ax.minorticks_on()

ax = plt.subplot(2,1,2)
ax.plot(freq, np.fft.fftshift(20*np.log10(abs(freq_response))))
ax.grid(which="major")
ax.grid(which="minor")
ax.minorticks_on()
ax.set_xlim(-620, 620)
ax.set_ylim(-100, 1)
```
