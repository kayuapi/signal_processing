```python
# Copyright Tay Zhi Cong, year 2025

import random
import numpy as np 
import matplotlib.pyplot as plt 
from scipy import signal 
import math
import sys
np.set_printoptions(threshold=sys.maxsize)
```

#### form a 22-tap low-pass half-band filter

```python
n = np.arange(-5.25, 5.5, 0.5) 
h = np.sinc(n)
```

#### stem the impulse response, plot the magnitude spectrum and then the log magnitude spectrum

```python
ax1 = plt.subplot(3,1,1)
ax1.stem(h)
ff = np.arange(0, 1, 1/1024) * 1
ax2 = plt.subplot(3,1,2)
ax2.plot(ff, abs(np.fft.fft(h, 1024)))
ax2.grid(True)
ax2.minorticks_on()
ax2.grid(which="minor")
ax2 = plt.subplot(3,1,3)
ax2.plot(ff, 10*np.log10(abs(np.fft.fft(h, 1024))))
ax2.grid(True)
ax2.minorticks_on()
ax2.grid(which="minor")
```

#### DC gain of the filter is approximately 2

#### Sum the impulse response to get the appropriate scale factor to apply to the impulse response to set the DC gain to unity

```python
print("sum of the impulse response is", sum(h))
```

#### the locations of the impulse response zeros do not exist. They sit betwen impulse response sammple 1 and sample 2, sample 3 and smample 4, sample 5 and sample 6, sample 7 and sample 8...

#### Scale the impulse response and replot it

```python
h = h/sum(h)
ax1 = plt.subplot(3,1,1)
ax1.stem(h)
ff = np.arange(0, 1, 1/1024) * 1
ax2 = plt.subplot(3,1,2)
ax2.plot(ff, abs(np.fft.fft(h, 1024)))
ax2.grid(True)
ax2.minorticks_on()
ax2.grid(which="minor")
ax2 = plt.subplot(3,1,3)
ax2.plot(ff, 10*np.log10(abs(np.fft.fft(h, 1024))))
ax2.grid(True)
ax2.minorticks_on()
ax2.grid(which="minor")
```

####  the length of the filter to have alternate values of the impulse response be zero

To have alternate values of the impulse response be zero, the length of the filter must be odd.

Note that having alternate values of the impulse response be zero is a desired property for a half-band filter and for a polyphase partition.
