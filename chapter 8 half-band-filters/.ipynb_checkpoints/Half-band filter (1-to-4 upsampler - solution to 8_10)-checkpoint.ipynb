```python
# Copyright Tay Zhi Cong, year 2025

import random
import numpy as np 
import matplotlib.pyplot as plt 
from scipy import signal 
import math
import sys
np.set_printoptions(threshold=sys.maxsize)
```

#### Design a low-pass half-band filter with the following spec:
transition bandwidth: 0.1 \
stop-band attenuation: 60dB

```python
stop_band_attenuation = 60
f_sampling = 1
f_transition = 0.1
f_minus_6db = 0.25

beta = signal.kaiser_beta(stop_band_attenuation) # 1. calculate the kaiser beta
filter_length = math.ceil((f_sampling/f_transition) * (stop_band_attenuation-8)/14) # 2. estimate the filter length : N ~= f_sampling/ f_transition * (Attenuation (in dB) - 8) / 14
print("beta is", beta)
print("filter_length is", filter_length)

filter_length = 41

h = signal.remez(20, [0, (0.25-0.05)*2, 0.5, 0.5], [1, 0], weight=[1, 1])
h_complete = np.zeros(41)
h_complete[1::2] = 0.5 * h
h_complete[20] = 0.5
h = h_complete
print(h)
print("sum of h is", sum(h))
# not needed
# h = h/sum(h)
```

#### stem the impulse response, plot the magnitude spectrum and then the log magnitude spectrum

```python
ax1 = plt.subplot(3,1,1)
ax1.stem(h)
ff = np.arange(0, 1, 1/1024) * 1
ax2 = plt.subplot(3,1,2)
ax2.plot(ff, abs(np.fft.fft(h, 1024)))
ax2.grid(True)
ax2.minorticks_on()
ax2.grid(which="minor")
ax2 = plt.subplot(3,1,3)
ax2.plot(ff, 20*np.log10(abs(np.fft.fft(h, 1024))))
ax2.grid(True)
ax2.minorticks_on()
ax2.grid(which="minor")

plt.figure()
ax3 = plt.subplot(2,1,1)
ax3.plot(ff, 20*np.log10(abs(np.fft.fft(h, 1024))))
ax3.grid(True)
ax3.minorticks_on()
ax3.grid(which="minor")
ax3.set_xlim(0, 0.25)
ax3.set_ylim(-0.01, 0.01)

ax4 = plt.subplot(2,1,2)
ax4.plot(ff, 20*np.log10(abs(np.fft.fft(h, 1024))))
ax4.grid(True)
ax4.minorticks_on()
ax4.grid(which="minor")
ax4.set_xlim(0.28, 0.33)
ax4.set_ylim(-80, -50)
```

```python
def generate_sinusoid(number_of_samples):
    n = np.arange(0, number_of_samples, 1)
    sinusoid = np.sin(2*np.pi*0*n) + np.sin(2*np.pi*0.05*n) + np.sin(2*np.pi*0.11*n) + np.sin(2*np.pi*0.155*n) + np.sin(2*np.pi*0.21*n) + np.sin(2*np.pi*0.26*n)
    return sinusoid

input = generate_sinusoid(200)
zero_packed_input = np.zeros(400)
zero_packed_input[::2] = input

first_stage_output = np.convolve(zero_packed_input, h)
plt.figure()
plt.plot(first_stage_output)
plt.figure(figsize=(10,8))
ax1 = plt.subplot(2,1,1)
ff = np.arange(0, 1, 1/1024)
ax1.plot(ff, 20*np.log10(abs(1/len(input)*np.fft.fft(input, 1024))))
ax1.set_ylim(-40,0)
ax1.grid(True)
ax1.minorticks_on()
ax1.grid(which="minor")
# ax1.set_xlim(0, 0.3)
ax1.set_title("Spectrum of input")

print("first_stage_output shape", first_stage_output.shape)

ff = np.arange(0, 1, 1/1024) * 2
ax2 = plt.subplot(2,1,2)
ax2.plot(ff, 20*np.log10(abs(1/len(first_stage_output)*np.fft.fft(first_stage_output, 1024))))
ax2.set_ylim(-40,0)
ax2.grid(True)
ax2.minorticks_on()
ax2.grid(which="minor")
ax2.set_xlim(0, 0.4)
ax2.set_title("Spectrum of output")

```
